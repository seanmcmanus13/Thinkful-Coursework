# Sorting Algos

```python
import time
import random

# Set seed.
random.seed(a=200)

# Create our default list.
short_list = list(random.sample(range(1000000), 10))
long_list = list(random.sample(range(1000000), 10000))
```

# Quick Sort

```python
#If given indexable and slicable iterable will return a sorted list
def quicksorter(list):

    if list: # if given list (or tuple) with one ordered item or more: 
        pivot = list[0]
        # below will be less than:
        below = [i for i in list[1:] if i < pivot] 
        # above will be greater than or equal to:
        above = [i for i in list[1:] if i >= pivot]
        return quicksorter(below) + [pivot] + quicksorter(above)
    else: 
        return list # empty list
```

```python
quicksorter(long_list)
```

# Selection Sort

```python
def selectionsort(list):
    #for all elements in list
    for i in range(len(list)): 
      
        # Find the minimum element in remaining  
        # unsorted array 
        willbemin = i 
        for j in range(i+1, len(list)): 
            if list[willbemin] > list[j]: 
                willbemin = j 
              
        # Swap the found minimum element with  
        # the first element         
        list[i], list[willbemin] = list[willbemin], list[i] 
        return list
```

```python
selectionsort(short_list)
```
