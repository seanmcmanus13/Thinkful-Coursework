```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
#for later use - ignore for now
from sklearn import datasets, linear_model
from sklearn.metrics import mean_squared_error, r2_score
```

```python
#importing dataset
UN_health_economics_df = pd.read_csv('Seans_UN_dataset.csv')
print(UN_health_economics_df)
```

```python
#Here's an example of a sub-plot:

plt.figure(figsize=(10,5))

plt.subplot(1, 2, 1)

# Plot them as a histogram.
plt.hist(UN_health_economics_df['life_expectancy']) 

plt.ylabel('Life Expectancy at birth (in years)')
plt.title('Histogram of Life Expectancy at birth')
plt.show()

plt.subplot(1, 1, 1)
plt.hist(UN_health_economics_df['per_capita_gov_expense_healthcare_2011'], bins=500)

plt.ylabel('Per capita government expenditure on healthcare 2011')
plt.title('Histogram of Government expenditure on healthcare 2011')
plt.show()


plt.tight_layout()
plt.show()
```

I'm interested in the relationships between life expectancy, healthcare outcomes and the modifable predictors of these metrics. The histograms below show that there are a few countries that relatively outperform others in terms of life expectancy. There are also a few that drastically underperform. 

My first guess would have been that healthcare expenditure would have reflected this data but this does not seem to be the case. There are a few big spenders who pump alot more money into healthcare than others. I would imagine these big spenders reflect the country's that are outliers for life expectancy, generally speaking.

```python
plt.scatter(x= UN_health_economics_df['per_capita_gov_expense_healthcare_2011'], y= UN_health_economics_df['GDP'])
plt.ylabel('Gross Domestic Product')
plt.xlabel('Per Capita Government Expenditure on Healthcare (2011)')
plt.title('The relationship between GDP and Government Healthcare expenditure')
plt.show()


plt.boxplot(UN_health_economics_df['per_capita_gov_expense_healthcare_2011'])
plt.show()
```

There seems to be linear relationship between GDP and Per Capita Expenditure Healthcare. However, the graph is suggestive of the prior established hypothesis that there are some notable outliers. 

Boxplot confirms there are some major outliers. There are quite a few fliers that are beyond 150% of the interquartile range. It seems to be that most countries spend relatively little on healthcare, when compared to a smaller number of relatively bigger spenders. It would be interesting to explore the relationships between GDP, taxation policy, healthcare outcomes and per capita economic productivity. This data is alligns with what was observed in the previously discussed histogram.

```python
plt.scatter(x= UN_health_economics_df['life_expectancy'], y= UN_health_economics_df['per_capita_gov_expense_healthcare_2011'])
plt.ylabel('LOG of Gross Domestic Product')
plt.xlabel('Life expectancy at birth')
plt.title('The relationship between GDP and Life Expectancy')
plt.yscale('log')
plt.show()
```

Given the large range of GDP and high variance, I opted to plot the log(GDP) against Life Expectancy at Birth. As expected, there's a clear relationship but also some clear outliers. I would like to see what the other readily modifiable strong predictors are and in particular how these predictors affect those countries that are clear outliers.
