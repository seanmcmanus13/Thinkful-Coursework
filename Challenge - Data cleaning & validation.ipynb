## Goal 2:
Next, calculate the mean, median, and standard deviation of the open-access cost per article for each journal . 

```python
#need to clean cost data

wellcome_df['cost_edit']=wellcome_df.COST.str.replace("$", "")
wellcome_df['cost_edit']=wellcome_df.cost_edit.str.replace("£", "")
wellcome_df['cost_edit']=wellcome_df.cost_edit.str.replace("€", "")

wellcome_df['cost_edit']=wellcome_df['cost_edit'].apply(pd.to_numeric)

                          
```

```python
#get mean (not accounting for currency conversions here...that also needs to be done...)
print('this is the mean data')
print(wellcome_df.groupby('journal_title_lower')['cost_edit'].mean())
```

```python
#get median (not accounting for currency conversions here...that also needs to be done...)
print('this is the median data')
print(wellcome_df.groupby('journal_title_lower')['cost_edit'].median()))
```

```python
#get standard deviation (not accounting for currency conversions here...that also needs to be done...)
print('these are the standard deviations')
print(wellcome_df.groupby('journal_title_lower')['cost_edit'].std())
```
